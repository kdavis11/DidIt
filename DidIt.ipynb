```python
#User Class
class User:
    def __init__(self, name, points):
      self.name = name
      self.tasks = []
      self.points = points
    #Getters

    def getName(self):
      return self.name

    def getPoints(self):
      return self.points

    def getTasks(self):
      return self.tasks
    #Setters

    def setName(self, name):
      self.name = name

    def setPoints(self, points):
      self.points = points

    #Task Methods
    def addTask(self, task):
      self.tasks.append(task)

    def removeTask(self, taskIndex):
      self.tasks.pop(taskIndex)

    def displayTasks(self):
      count = 1
      for task in self.tasks:
        print(str(count) + ". " + task.getTitle())
        count += 1

    def addPoints(self, points):
      self.points += points

    def subtractPoints(self, points):
      self.points -= points

    def displayMenu(self):
      print("Hello, " + self.name + "!")
      print("Please select an option: ")
      print("1. Add a new task")
      print("2. Change a task's status")
      print("3. Delete a task")
      print("4. Buy a decoration")
      print("5. Change decorating style")
      print("6. View tasks")

#Task Class
class Task:
    def __init__(self, title, description, status, dueDate):
      self.title = title
      self.description = description
      self.status = status
      self.dueDate = dueDate

    #Getters

    def getTitle(self):
      return self.title

    def getDescription(self):
      return self.description

    def getStatus(self):
      return self.status

    def getDueDate(self):
      return self.dueDate

    #Setters

    def setTitle(self, title):
      self.title = title

    def setDescription(self, description):
      self.description = description

    def setStatus(self, status):
      self.status = status

    def setDueDate(self, dueDate):
      self.dueDate = dueDate


```

```python
print("Welcome to DidIt! To get started, please enter your name: ")
username = input()
p1 = User(username, 0)
choice = 0
owned = [] #Tracks the decorations the user owns
equipped = "" #The string stored here determines which decoration is displayed, if any.
while True:
    #Currently displays 30 of each symbol
    if equipped == "Stars":
      print("*******************************")
      p1.displayMenu()
      print("*******************************")
    elif equipped == "Hearts":
      print("♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥")
      p1.displayMenu()
      print("♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥♥")
    else:
      p1.displayMenu()

    choice = int(input())
    while int(choice) < 1 or int(choice) > 6: #Error trapping
      print("Invalid choice! Please try again: ")
      choice = int(input())

    if choice == 1: #Add task
      taskName = ""
      taskDesc = ""
      taskStatus = ""
      taskDueDate = ""

      print("Please enter the name of the task: ")
      taskName = input()
      print("Please enter a description of the task: ")
      taskDesc = input()
      print("Choose a status for the task: ")
      print("1. DoIt")
      print("2. DoingIt")
      choice = int(input())
      if choice == 1:
        taskStatus = "DoIt"
      elif choice == 2:
        taskStatus = "DoingIt"
      print("Please enter a due date for the task: ")
      taskDueDate = str(input())
      t = Task(taskName, taskDesc, taskStatus, taskDueDate)
      p1.addTask(t)
      print("Task successfully added!")

    elif choice == 2: #Change a task's status
      print("Please select the task you would like to change: ")
      p1.displayTasks()
      choice = int(input())
      while choice - 1 < 0 or choice - 1 > len(p1.tasks): #Error trapping
        print("Invalid task number! Please try again: ")
        choice = int(input())
      print("Choose a status for the task: ")
      print("1. DoIt")
      print("2. DoingIt")
      print("3. DidIt")
      statusChoice = int(input())
      if statusChoice == 1:
        p1.tasks[choice - 1].setStatus("DoIt")
      elif statusChoice == 2:
        p1.tasks[choice - 1].setStatus("DoingIt")
      elif statusChoice == 3:
        p1.tasks[choice - 1].setStatus("DidIt")
        p1.addPoints(10)
      print("Task status successfully changed!")

    elif choice == 3: #Delete a task
      print("Please select the task you would like to delete: ")
      p1.displayTasks()
      choice = int(input())

      while choice - 1 < 0 or choice - 1 > len(p1.tasks): #Error trapping
        print("Invalid task number! Please try again: ")
        choice = int(input())

      p1.removeTask(choice - 1)
      print("Task successfully deleted!")


    elif choice == 4: #Buy a decoration
      print("Welcome to the Decoration Shop!")
      print("You have " + str(p1.getPoints()) + " points!")
      print("Please select one decoration")
      decorations = {"Stars": 10, "Hearts": 10}
      for i, (item, price) in enumerate(decorations.items(), start = 1):
        print(f"{i}. {item} - {price} points")

      buyChoice = int(input("Enter the number of the decoration you want to buy: "))

      while buyChoice < 1 or buyChoice - 1 > len(decorations): #Error trapping
        print("Invalid choice! Please try again. ")
        buyChoice = int(input("Enter the number of the decoration you want to buy: "))

      names = list(decorations.keys())
      name = names[buyChoice - 1]
      cost = decorations[name]
      if p1.getPoints() >= cost:
            p1.subtractPoints(cost)
            item_name = list(decorations.keys())[buyChoice - 1] #Takes all of the keys in the dictionary and puts them into a list. Then, it stores the item name corresponding to the number entered by the user.
            owned.append(item_name) #The above line was necessary to be able to write this line, which adds the bought item into the owned list
            print(f"You have successfully bought {item_name}! You can equip it from the menu.")
            decorations.pop(names[buyChoice-1])

      else:
        print("You don’t have enough points to buy that!")

    elif choice == 5: #Change decorating style
      print("Please select a decoration to equip: ")
      for i in range(len(owned)):
        print(f"{i+1}. {owned[i]}")
      print(f"{len(owned)+1}. Unequip all decorating styles")
      equipChoice = int(input())

      if equipChoice != len(owned) + 1:
        equipped = owned[equipChoice - 1]
        print("You have successfully equipped the " + equipped + " decoration style!")
      else:
        equipped = ""
        print("You have successfully unequipped your decorating style!")

    elif choice == 6: #View tasks
      count = 1
      for task in p1.getTasks():
        print(str(count) + ". " + task.getTitle() + " Description: " + task.getDescription() + " Status: " + task.getStatus() + " Due Date: " + task.getDueDate())
        count += 1

```

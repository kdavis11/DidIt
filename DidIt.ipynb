```python
#User Class
class User:
    def __init__(self, userID, name, points):
      self.userID = userID #Might be able to remove this because I don't think we're expected to actually create a legitimate app. If it's confirmed we won't make a real app, then we can delete this
      self.name = name
      self.tasks = []
      self.points = points
    #Getters
    def getUserID(self):
      return self.userID

    def getName(self):
      return self.name

    def getPoints(self):
      return self.points

    #Setters
    def setUserID(self, userID):
      self.userID = userID

    def setName(self, name):
      self.name = name

    def setPoints(self, points):
      self.points = points

    #Task Methods
    def addTask(self, task):
      self.tasks.append(task)

    def removeTask(self, taskIndex):
      self.tasks.pop(taskIndex)

    def displayTasks(self):
      count = 1
      for task in self.tasks:
        print(str(count) + ". " + task.getTitle())
        count += 1

    def addPoints(self, points):
      self.points += points

    def subtractPoints(self, points):
      self.points -= points

#Task Class
class Task:
    def __init__(self, taskID, userID, title, description, status, dueDate): #Might be able to remove the IDs here as well depending on if we need to create an actual app or not
      self.taskID = taskID
      self.userID = userID
      self.title = title
      self.description = description
      self.status = status
      self.dueDate = dueDate

    #Getters
    def getTaskID(self):
      return self.taskID

    def getUserID(self):
      return self.userID

    def getTitle(self):
      return self.title

    def getDescription(self):
      return self.description

    def getStatus(self):
      return self.status

    def getDueDate(self):
      return self.dueDate

    #Setters
    def setTaskID(self, taskID):
      self.taskID = taskID

    def setUserID(self, userID):
      self.userID = userID

    def setTitle(self, title):
      self.title = title

    def setDescription(self, description):
      self.description = description

    def setStatus(self, status):
      self.status = status

    def setDueDate(self, dueDate):
      self.dueDate = dueDate


```

```python
print("Welcome to DidIt! To get started, please enter your name: ")
username = input()
p1 = User(0, username, 0) #Just using 0 for the user ID because we might be able to remove the IDs entirely later, and initializing points to 0
choice = 0
while True:
    print("Hello, " + p1.getName() + "!")
    print("Please select an option: ")
    print("1. Add a new task")
    print("2. Change a task's status")
    print("3. Delete a task")
    print("4. Buy a decoration")
    print("5. Change decorating style")
    choice = int(input())
    while int(choice) < 1 or int(choice) > 4: #Error trapping
      print("Invalid choice! Please try again: ")
      choice = int(input())

    if choice == 1: #Add task
      taskName = ""
      taskDesc = ""
      taskStatus = ""
      taskDueDate = ""

      print("Please enter the name of the task: ")
      taskName = input()
      print("Please enter a description of the task: ")
      taskDesc = input()
      print("Choose a status for the task: ")
      print("1. DoIt")
      print("2. DoingIt")
      choice = int(input())
      if choice == 1:
        taskStatus = "DoIt"
      elif choice == 2:
        taskStatus = "DoingIt"
      t = Task(0, p1.getUserID(), taskName, taskDesc, taskStatus, taskDueDate)
      p1.addTask(t)
      print("Task successfully added!")

    elif choice == 2: #Change a task's status
      print("Please select the task you would like to change: ")
      p1.displayTasks()
      choice = int(input())
      while choice - 1 < 0 or choice - 1 > len(p1.tasks): #Error trapping
        print("Invalid task number! Please try again: ")
        choice = int(input())
      print("Choose a status for the task: ")
      print("1. DoIt")
      print("2. DoingIt")
      print("3. DidIt")
      statusChoice = int(input())
      if statusChoice == 1:
        p1.tasks[choice - 1].setStatus("DoIt")
      elif statusChoice == 2:
        p1.tasks[choice - 1].setStatus("DoingIt")
      elif statusChoice == 3:
        p1.tasks[choice - 1].setStatus("DidIt")
        p1.addPoints(10)
      print("Task status successfully changed!")

    elif choice == 3: #Delete a task
      print("Please select the task you would like to delete: ")
      p1.displayTasks()
      choice = int(input())
      while choice - 1 < 0 or choice - 1 > len(p1.tasks): #Error trapping
        print("Invalid task number! Please try again: ")
        choice = int(input())
      p1.removeTask(choice - 1)
      print("Task successfully deleted!")

    elif choice == 4: #Buy a decoration
      print("Welcome to the Decoration Shop!")
      print("You have " + str(p1.getPoints()) + " points!")
      print("Please select one decoration")
      decorations = {"1":("Stars", 10), "2":("Hearts", 10)}
      for key, value in decorations.items():
        print(f"{key}. {value[0]} - {value[1]} points")
      buyChoice = input("Enter the number of the decoration you want to buy: ")
      if buyChoice in decorations:
        name, cost = decorations[buyChoice]
        if p1.getPoints() >= cost:
            p1.addPoints(-cost)
        else:
            print("You don’t have enough points to buy that!")
      else:
        print("Invalid choice!")

    elif choice == 5: #Change decorating style
      print("")
```
